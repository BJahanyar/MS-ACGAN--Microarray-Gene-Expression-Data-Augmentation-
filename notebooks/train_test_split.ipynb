```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataset/8_202.csv', index_col=0, low_memory=False)
```

```python
train_df, test_df = train_test_split(df.T, shuffle=True, test_size=0.25)
```

```python
train_df.T.to_csv('dataset/8_202_train.csv')
test_df.T.to_csv('dataset/8_202_test.csv')
```
